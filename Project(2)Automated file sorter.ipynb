# Automated File Sorter

Automatically organises files in a directory into subfolders 
by file type (CSV, image, text).

**Libraries:** os, shutil  
**Skills demonstrated:** file system automation, conditional logic, 
directory management

```python
import os, shutil

# Set this to the folder you want to sort
# Change this path to match your own directory
path = r"C:/Your/Folder/Path/Here/"
```

```python
file_name=os.listdir(path)
```

```python
folder_name=['csv files','image files','text files']

for loop in range(0,3):
    if not os.path.exists(path + folder_name[loop]):
        #print(path + folder_name[loop])
        os.makedirs(path + folder_name[loop])
for file in file_name:
    if '.csv' in file and not os.path.exists(path + 'csv files/' + file):
        shutil.move(path + file, path +'csv files/' + file)
    elif '.jpg' in file and not os.path.exists(path + 'image files/' + file):
        shutil.move(path + file, path +'image files/' + file)
    elif '.txt' in file and not os.path.exists(path + 'text files/' + file):
        shutil.move(path + file, path +'text files/' + file)
```
